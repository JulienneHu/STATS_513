##### 4. Compute the sample correlation between the residuals and the fit-ted values. Plot residuals against fitted values. Explain the value of this correlation using the geometric (projection) interpretation of least squares.


```r
cor(residuals(regre_model), fitted(regre_model))
plot(regre_model, which = 1)
```

The sample correlation between the residuals and the fitted value is -1.14138128387761e-16.\
In definition, regression will give a linear trend based on predicators and residuals are the randomness that is “leftover from a regression model” from fitting a regression model. The correlation between the explanatory variable(s) and the residuals is/are zero because there’s no linear trend left - it’s been removed by the regression.\
![image.png](HW1.png)
According to vector space interception, $min_{\beta}(y-X\beta)^T(y-X\beta)$ minimizes the Euclidean distance between $y$ and the linear space spanned by the columns of X. Here the fitted data is $X\beta$, so the residuals are perpendicular to the fitted data. Hence, the correlation between the residuals and the fitted value is almost 0.
